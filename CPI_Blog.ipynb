```python
# Import libraries

# Open websites using urllib
import urllib2
# import beautifulsoup
from bs4 import BeautifulSoup
# import os to set working directory
import os
```

```python
# Set working directory
os.chdir('C:\\Users\\Jasper\\Desktop\\')

# define URL (CPI 2006)
url = "http://archive.transparency.org/policy_research/surveys_indices/cpi/2006"

# Open page and beautifulsoup page
page = urllib2.urlopen(url)
soup = BeautifulSoup(page)

# Look at HTML code
#print(soup.prettify())
```

```python
# Find the correct table. Method #1 (long way)
tables = soup.findAll('table')
print(tables[3])

# Find the correct table. Method #2 (shortcut)
CPItab = soup.find('table',class_='integrity_awards')
```

```python
# Extract all <tr> tags
CPI2006 = CPItab.findAll('tr')
print(CPI2006[0])
```

```python
# Open connection to a .txt file and create column headers
f = open('CPI2006.txt', 'w')
f.write("Rank" + ';' + "Country" + ';' + "CPI_score" + ';' + "Number_surveys" + ';' + "Conf_Range_low" + ';' + "Conf_Range_high" + ";" + "Conf_Range_Difference" + '\n')

# Loop
for i in CPI2006:
    # Isolate the <p>-tags
    ptags = i.findAll('p')
    # Store values in objects
    Rank = ptags[0].contents[0].replace("\n", "")
    Country = ptags[1].contents[0].replace("\n", "")
    CPIScore =  ptags[2].contents[0].replace("\n", "")
    NoSurv = ptags[3].contents[0].replace("\n", "")
    ConfRange_low = ptags[4].contents[0].replace("\n", "").split('-')[0]
    ConfRange_high = ptags[4].contents[0].replace("\n", "").split('-')[1]
    Conf_Difference = float(ConfRange_high) - float(ConfRange_low)
    
    # Write to file
    
    f.write(str(Rank) + ';' + str(Country) + ';' + str(CPIScore) + ';' + str(NoSurv) + ';' + str(ConfRange_low) + ';' + str(ConfRange_high) + ';' + str(Conf_Difference) + '\n')
    
f.close()
```
